## Generative Adversarial Networks (GANs)

Generative Adversarial Networks (GANs) tabiri Üretken Çekişmeli Ağlar anlamına gelmektedir. Bu tabirden anlaşılacağı üzere ağlar birbirleriyle çekişme halindedirler.

Kalpazan (Generator) bir tane resim yapıyor ve bu resmin gerçek olduğunu söylüyor. İki input olan gerçek data (Real Data) ve feyk data (Generator) detektife geliyor. Dedektif gelen resimlerden hangisinin gerçek hangisinin sahte olduğunu anlamaya çalışıyor. Bu arada kalpazan giderek gerçek resim yapmayı öğreniyor. Böylece eğitim setiyle yani gerçek data ile aynı istatistiğe sahip yeni veriler oluşturmayı öğreniyor. 

Tüm bu anlatılanlar göz önünde bulundurularak GAN, eğitim seti ile aynı istatistiklere sahip yeni veriler oluşturmayı öğrenen yapılardır.

GAN 2 tane network'ten oluştuğu için genelde sonuna 's' koyularak GANs denir.

2 tane network var ve bu network'lerden bahsedelim:

    Generator Network: Üretici ağıdır.
    
    Descriminator Network: Ayırt edici ağdır. 
    
Üretici ağ, bir input var ve bunun sonucunda da bir output veriyor. Yani yeni bir şey üretiyor.

Ayırt edici ağ, deep learning ve machine learning algoritmalarında gördüğümüz birnevi classification yöntemidir. Yani ayırt ediyo, sınıflandırıyo diyebiliriz.

Generator'e kalpazan diyorlar. Kalpazan, sahte şeyler üreten biridir. 

Descriminator'e Dedektif diyorlar. Ayırt etme işlemi yaptığı için dedektif diyorlar.

Generator'ün yani üretici ağın amacı sürekli yeni data üretmektir. Buradaki asıl amaç gerçeğinden ayırt edilemez data'lar üretmektir.
GAN'ın tanımını hatırlayalım. Eğitim setiyle aynı istatistiklere sahip yeni veriler üretmektir.

Amacımız yeni veri üretmektir. Fakat mesela bir kedi üreteceksek gerçeğe yakın bir kedi üretilmesi gerekmektedir. Bu üretimi Generator yapacak. Başlangıçta saçma sapan resimler üretecektir ve bu resimleri zamanla gerçek haline benzetmeye çalışacaktır. Dolayısıyla ilk öğrenecek olan üretici ağlardır. Yani Generator ağı bu yapıyı öğrenmede kullanılacak ilk yapıdır.

Descriminator ise ayırt edici ağdır. Generator bir şeyler üretecek ve discriminator'de üretilen data gerçek mi değil mi bunu anlamak zorundadır. Yani burada da bir öğrenme yöntemi söz konusudur. Yani GAN yapısında veriden bir şeyler öğrenecek ikinci network'te descriminator network'tür.

Görseli açıklayayım. Bir eğitim seti vardır. Eğitim setinden yeni veri üretilecektir. Generator isimli neural network yapısı yeni veri üretecektir.

Eğitim seti, gerçek veridir. Kutu içerisindeki Real sample, gerçek veridir. Bu gerçek veriyi label'layacağız. Bu verinin label'ı real olucak. Ya da real yerine y=1 şeklinde de labellayabiliriz. Bunu dedektif olan Descriminator'e veriyoruz. Dedektife diyoruz ki, elimizde bir real sample var ve bunu y=1 etiketi ile sana verdim. Sonrasında descriminator real ve feyk olanı ayırt ediyor. Demek ki descriminator'e birde feyk data gelecek. Feyk data'nın label'ı y=0 olsun. Feyk data, üretilen data'dır ve bu data generator'den geliyor. Generator input olarak random noise alıyor ve data üretiyor. Random noise, rastgele gürültü demektir. Gürültülü bir veriyi alıyor, neural network yapısı bir resim yani data üretiyor ve üretilen data feyk olarak label'lanıp descriminator'ün içerisine yollanır.

Sonrasında descriminator'ün real mı feyk mi olduğuna dair karar vermesi gereken bir süreç söz konusudur. Real mı feyk mi data olup olmadığına karar verilir ve sonra descriminator'e back propagation yapılır ve Generator'ede back propagation yapılır. Back propagation yöntemiyle generator daha gerçekçi resimler yapmayı öğreniyor ve descriminator bunu daha doğru ayırt etmeyi öğreniyor. Bu ikisi arasında bir çekişme söz konusu ve zaten bu yüzden Generative Adversarial Networks deniyor.

Generator network'ü fake resimler üreterek descriminator'ü kandırmaya çalışıyor ve descriminator'de bu fake ve real resimleri öğrenerek hangi resmin gerçek hangi resmin sahte olduğunu öğreniyor ve bu Generator tarafından kandırılmayı önlemeye çalışıyor.

Generator gerçeğe yakın resimler ürettiği için Descriminator'ü kandırmaya başlayacak. Böylelikle real sample'a yani eğitim setine göre bu eğitim seti ile aynı istatistiklere sahip yeni veriler oluşturmayı öğrenecek. Artık bizim öğrenmiş bir generator network'ümüz olacak ve buna random noise verdiğimiz zaman gerçek resimlere çok yakın resimler elde etmiş olucaz. GAN'ın genel mantığı bu şekildedir. 

Şimdi Generator network'ün train sürecinin nasıl gerçekleştiğine bakalım.

Bir noise var ve bu noise'yi Generator network'ün içerisine veriyoruz. Ve bir resim elde ediliyor. Bu resim descriminator'e verilir. Burada önemli bir nokta resimleri label'ladığımızdır. Ama Generator'ün amacı Descriminator'ü kandırmak olduğu için fake resmin y=0 etiketi y=1 olur. Çünkü fake resmi gerçek resim diye Descriminator'ü kandırmak istiyoruz. Descriminator kanar ya da kanmaz ama Generator bu şekilde öğreniyor.

```python
from keras.layers import Dense, Dropout, Input, ReLU
from keras.models import Model, Sequential
from keras.optimizers import Adam
from keras.datasets import mnist
import numpy as np
import matplotlib.pyplot as plt
```

mnist veri seti 0'dan 9'a kadar olan sayılara ait resimler içerir. mnist veri setini yükleyelim.

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = (x_train.astype(np.float32)-127.5)/127.5 #Bu satır ile tüm değerler -1 ile 1 arasına sıkıştırılmıştır.
```

```python
print(x_train.shape)
```

Keras'ta 28,28 yerine 28x28=784 yazılmalıdır. Dolayısıyla reshape yapalım.

```python
x_train = x_train.reshape(x_train.shape[0],x_train.shape[1]*x_train.shape[2])
print(x_train.shape)
```

```python
plt.imshow(x_test[12])
```

Oluşturacağımız generator yapısı şu şekildedir:

<br>
<img src="g3.png" style="width:800px;height:300px"/> 
Şimdi bu yapıyı oluşturalım.

```python
# create generator
def create_generator():
    
    generator = Sequential()
    generator.add(Dense(units = 512, input_dim = 100))
    generator.add(ReLU())
    
    generator.add(Dense(units = 512))
    generator.add(ReLU())
    
    generator.add(Dense(units = 1024))
    generator.add(ReLU())
    
    generator.add(Dense(units = 784, activation = "tanh"))
    
    generator.compile(loss = "binary_crossentropy",
                      optimizer = Adam(lr = 0.0001, beta_1 = 0.5))
    return generator

g = create_generator()
g.summary()   
```

Oluşturacağımız descriminator yapısı şu şekildedir:

<br>
<img src="g4.png" style="width:800px;height:300px"/> 
Şimdi bu yapıyı oluşturalım.

```python
# descriminator

def create_discriminator():
    discriminator = Sequential()
    discriminator.add(Dense(units=1024,input_dim = 784))
    discriminator.add(ReLU())
    discriminator.add(Dropout(0.4))
    
    discriminator.add(Dense(units=512))
    discriminator.add(ReLU())
    discriminator.add(Dropout(0.4))
    
    discriminator.add(Dense(units=256))
    discriminator.add(ReLU())
    
    discriminator.add(Dense(units=1, activation = "sigmoid"))
    
    discriminator.compile(loss = "binary_crossentropy",
                          optimizer= Adam(lr = 0.0001, beta_1=0.5))
    return discriminator

d = create_discriminator()
d.summary()
```

Oluşturacağımız GANs yapısı şu şekildedir:

<br>
<img src="g5.png" style="width:500px;height:100px"/> 

Şimdi bu yapıyı oluşturalım.

```python
# gans
def create_gan(discriminator, generator):
    discriminator.trainable = False
    gan_input = Input(shape=(100,))
    x = generator(gan_input)
    gan_output = discriminator(x)
    gan = Model(inputs = gan_input, outputs = gan_output)
    gan.compile(loss = "binary_crossentropy", optimizer="adam")
    return gan
    
gan = create_gan(d,g)
gan.summary()
```

```python
#train

epochs = 50
batch_size = 256

for e in range(epochs):
    for _ in range(batch_size):
        
        noise = np.random.normal(0,1, [batch_size,100])
        
        generated_images = g.predict(noise)
        
        image_batch = x_train[np.random.randint(low = 0, high = x_train.shape[0],size = batch_size)]
        
        x = np.concatenate([image_batch, generated_images])
        
        y_dis = np.zeros(batch_size*2)
        y_dis[:batch_size] = 1
        
        d.trainable = True  #Eğitime başlandı.
        d.train_on_batch(x,y_dis)

        noise = np.random.normal(0,1,[batch_size,100])
        
        y_gen = np.ones(batch_size)
        
        d.trainable = False  #Artık generator'ü eğiteceğim için discriminator'ün eğitimini False'e eşitledim.
        
        gan.train_on_batch(noise, y_gen)
    print("epochs: ",e)
```

```python
# save model
g.save_weights('gans_model.h5')  # always save your weights after training or during training
```
